```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '###!', '####', '****', '*****', 'NULL']
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True) #sep - seprador, sep=";" parse_dates (atributo usado para ele entender qual coluna vai ser transformada ou interpretada do tipo data). dayfirst=True define que o dia vem primeiro, ficnado no seguinte formato d/m/a.
df.head()
```

```python
schema = pa.DataFrameSchema(columns={
                                     "codigo":pa.Column(pa.Int, required=False), # Valida ou não uma coluna, pelo o required
                                     "codigo_ocorrencia":pa.Column(pa.Int), # No primeiro exemplo, define um schema usando a biblioteca pandera, definindo na coluna o tipo de valores, tipo int
                                     "codigo_ocorrencia2":pa.Column(pa.Int),
                                     "ocorrencia_classificacao":pa.Column(pa.String),
                                     "ocorrencia_cidade":pa.Column(pa.String),
                                     "ocorrencia_uf":pa.Column(pa.String, nullable=True), #pa.Check.str_length(2,2)), # Checa o tamanho mínimo de caracteres e o tamanho máximo
                                     "ocorrencia_aerodromo":pa.Column(pa.String),
                                     "ocorrencia_dia":pa.Column(pa.DateTime),
                                     "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), # nullable=True (permite valores nulos), expressão regular para hora, minuto e segundos
                                     "total_recomendacoes":pa.Column(pa.Int)}) 
```

```python
schema.validate(df) # Valida o schema no dataframe
```

```python
#ocorrencias cuja classificacao == INCIDENTE GRAVE
filtro = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro]
```

```python
#Combinação de filtros, cuja classificação == INCIDENTE GRAVE e o estado == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#Combinação de filtros, cuja classificação == INCIDENTE GRAVE ou estado == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 | filtro2]  # | = ou
```

```python
#Combinação de filtros, cuja classificação == INCIDENTE GRAVE ou classificação == INCIDENTE e o  estado == SP
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE')  | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2  ]  # | = ou
```

```python
#Combinação de filtros, cuja classificação == INCIDENTE GRAVE ou classificação == INCIDENTE e o  estado == SP, utilizando isin
filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2  ]  # | = ou
```

```python
# Todas as ocorrencias cuja a ocorrencia_cidade comece com a letra C
filtro = df.ocorrencia_cidade.str[0] == 'C' # Localiza o primeiro indice da string e verifica se é igual a 'C'
df.loc[filtro]
```

```python
# Todas as ocorrencias cuja a ocorrencia_cidade terminam com a letra A
filtro = df.ocorrencia_cidade.str[-1] == 'A' # Localiza o ultimo indice da string e verifica se é igual a 'A'
df.loc[filtro]
```

```python
# Todas as ocorrencias cuja a ocorrencia_cidade terminam com MA
filtro = df.ocorrencia_cidade.str[-2:] == 'MA' # Do penultimo em diante, verifique se é igual a 'MA'
df.loc[filtro]
```

```python
# Todas as ocorrencias cuja a ocorrencia_cidade contém o MA, usando o método 'contains'
filtro = df.ocorrencia_cidade.str.contains('MA') # Verifica se contém o 'MA'
df.loc[filtro]
```

```python
# Todas as ocorrencias cuja a ocorrencia_cidade contém o MA ou AL, usando o método 'contains'
filtro = df.ocorrencia_cidade.str.contains('MA|AL') # Verifica se contém o 'MA' ou 'AL
df.loc[filtro]
```

```python
# Ocorrencias do ano de 2015 e mes 12
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12)
df.loc[filtro]
```

```python
# Ocorrencias do ano de 2015 e mes 12, e no dia 8
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia = df.ocorrencia_dia.dt.day == 8
df.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
# Ocorrencias do ano de 2015 e mes 12, e no dia 3 ao dia 8
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia = (df.ocorrencia_dia.dt.day > 2) & (df.ocorrencia_dia.dt.day < 9)
df.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
# Ocorrencias do ano de 2015 e mes 12, e no dia 3 ao dia 8
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia_inicio = df.ocorrencia_dia.dt.day > 2
filtro_dia_fim = df.ocorrencia_dia.dt.day < 9
df.loc[filtro_ano & filtro_mes & filtro_dia_inicio & filtro_dia_fim]
```
