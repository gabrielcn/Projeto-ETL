```python
import pandas as pd
import pandera as pa
```

```python
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True) #sep - seprador, sep=";" parse_dates (atributo usado para ele entender qual coluna vai ser transformada ou interpretada do tipo data). dayfirst=True define que o dia vem primeiro, ficnado no seguinte formato d/m/a.
df.head(10)
```

```python
schema = pa.DataFrameSchema(columns={
                                     "codigo":pa.Column(pa.Int, required=False), # Valida ou não uma coluna, pelo o required
                                     "codigo_ocorrencia":pa.Column(pa.Int), # No primeiro exemplo, define um schema usando a biblioteca pandera, definindo na coluna o tipo de valores, tipo int
                                     "codigo_ocorrencia2":pa.Column(pa.Int),
                                     "ocorrencia_classificacao":pa.Column(pa.String),
                                     "ocorrencia_cidade":pa.Column(pa.String),
                                     "ocorrencia_uf":pa.Column(pa.String), #pa.Check.str_length(2,2)), # Checa o tamanho mínimo de caracteres e o tamanho máximo
                                     "ocorrencia_aerodromo":pa.Column(pa.String),
                                     "ocorrencia_dia":pa.Column(pa.DateTime),
                                     "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), # nullable=True (permite valores nulos), expressão regular para hora, minuto e segundos
                                     "total_recomendacoes":pa.Column(pa.Int)}) 
```

```python
schema.validate(df) # Valida o schema no dataframe
```
